```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
np.random.seed(2)
X = 2 * np.random.rand(100,1) # 100 points between 0 and 2
y = 3 + 2 * X + np.random.randn(100,1) # Linear relation with noise
```

```python
#Plot the data 
plt.scatter(X,y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Learning Data")
plt.show()
```

## Compute the cost function

```python
def compute_cost(X, y, theta0, theta1):
    m = len(X)
    predictions = theta0 + theta1 * X
    cost = (1/(2*m)) * np.sum((predictions - y)**2)
    return cost

print(compute_cost(X,y,3,2))
```
